## Nivel III (7.5 ptos)

Pregunta 8: Muestre la suma de cantidad de bienes utilizados vistos por departamento y provincia de los departamentos de Amazonas y San Martin y muestre ordenado de menor a mayor por la cantidad de bienes utilizados (tabla mobiliario) (2.5 ptos)

```python
# Desarrolle aqui
listDim = ['DEPARTAMENTO', 'PROVINCIA']
dictMetrica = {'CANTIDAD_BIEN':'sum'}
filtro = ((dfMobiliario['DEPARTAMENTO'] == 'Amazonas') | (dfMobiliario['DEPARTAMENTO'] == 'San Martin'))
dfMobiliario.loc[filtro].groupby(listDim, as_index = False).agg(dictMetrica).sort_values(by='CANTIDAD_BIEN', ascending=True)
```

```python
# Verificando que los resultados sean correctos
df1 = dfMobiliario.loc[dfMobiliario['DEPARTAMENTO'] == 'Amazonas']
print("Las provincias de Amazonas son:", df1['PROVINCIA'].unique())
df2 = dfMobiliario.loc[dfMobiliario['DEPARTAMENTO'] == 'San Martin']
print("Las provincias de San Martin son:", df2['PROVINCIA'].unique())
```

Pregunta 9: Muestre el promedio del PBI visto por region y muestre ordenado de mayor a menor quien tiene mas PBI (tabla Indicadores) (2 ptos)

```python
dfIndicadores.head(5)
```

```python
# Desarrolle aqui
listDim = ['Region']
dictMetrica = {'PBI':'mean'}
dfIndicadores.groupby(listDim, as_index = False).agg(dictMetrica).sort_values(by='PBI', ascending=False)
```
