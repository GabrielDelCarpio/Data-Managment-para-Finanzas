## Pregunta 3: Importar datas "empleados" y hacer 2 tipos de tratamientos distintos al campo Job_ID ( 3 ptos)

```python
path_archivo = 'empleados.xlsx'
df = pd.read_excel(path_archivo, header=0)
df.head()
```

```python
df['JOB_ID'].unique()
```

```python
# 1er tratamiento: separación por tipo de trabajo e identificación
df[['JOB', 'ID']] = df['JOB_ID'].str.split('_', expand=True)
df.head()
```

```python
# 2do tratamiento: contamos los tipos de JOB_ID
df['JOB_ID'].value_counts()
```

```python
# Los tipo de la parte JOB
df['JOB'].value_counts()
```

# Pregunta 5: Pasar a dataframe la data del siguiente API (4 ptos)

```python
url = 'https://pokeapi.co/api/v2/pokemon-habitat/'
```

```python
# Verificamos si la API esta bien ejecutada
response = requests.get(url)
print(type(response))
print(response.status_code)
response.text
```

```python
# Verificamos si es una lista para pasar a df
data = json.loads(response.text)
print(type(data))
data
```

```python
# Finalmente convertimos a data frame
df = pd.DataFrame(data['results'])
df
```
